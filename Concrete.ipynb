```python
import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from sklearn.metrics import mean_squared_error
```

```python
classifier= Sequential()
classifier.add(Dense(activation="relu", input_dim=8, units=50, kernel_initializer="he_uniform"))
classifier.add(Dense(activation="relu", input_dim=8, units=50, kernel_initializer="he_uniform"))
classifier.add(Dense(activation="sigmoid", input_dim=8, units=1, kernel_initializer="glorot_uniform"))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer='adam',
    loss='mean_squared_error',
    loss_weights=None,
    sample_weight_mode=None,
    weighted_metrics=None,
    target_tensors=None)
```

```python
classifier.fit(x=x_train,
    y=y_train,
    batch_size=32,
    epochs=100,
    verbose=1,
    callbacks=None,
    validation_split=0.0,
    validation_data=(x_test,y_test),
    shuffle=True,
    class_weight=None,
    sample_weight=None,
    initial_epoch=0,
    steps_per_epoch=None,
    validation_steps=None,
    validation_freq=1,
    max_queue_size=10,
    workers=1,
    use_multiprocessing=False,)
```

```python
y_predtrain=classifier.predict(x_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_predtrain)
```

```python
y_predtest=classifier.predict(x_test)
```
